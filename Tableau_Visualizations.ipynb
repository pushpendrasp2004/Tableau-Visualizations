Task 1: Category Performance Analysis

Business Need:
Management wants to know which product categories generate the highest revenue.

Task:
Create a visualization to compare total Sales by Category and identify the top-performing category.

- Visualization: Bar Chart (Category vs Total Sales)

Answer:

Create a bar chart with Category on X-axis and Sum of Sales on Y-axis.
Compare total sales across categories.

Insight (example):

The category with the highest sales is typically Technology / Furniture / Office Supplies (depends on dataset).
This category contributes the most revenue and should be prioritized for growth strategies.

Task 2: Sales Trend Analysis

Business Need:
The leadership team wants to understand how sales fluctuate over time to plan promotions and inventory.

Task:
Analyze Sales trends over time (monthly) to identify peak and slow periods.

- Visualization: Line Chart (Monthly Sales Trend)

Answer:

Use Order Date (Month-Year) on X-axis and Sales on Y-axis.
Analyze seasonal patterns.

Insight:

Sales usually peak during festive/end-of-year months.
Low sales periods indicate when promotions or discounts are needed.

Task:

Plot Region vs Sales contribution (%).

One region (e.g., West/East/North/South) contributes the highest share.

Task 4: Profitability Investigation

Business Need:
The finance team suspects that some high-sales orders may be reducing overall profit.

Task:
Analyze the relationship between Sales and Profit to identify loss-making or low-profit orders.

-  Visualization: Scatter Plot (Sales vs Profit)

Answer:

X-axis: Sales
Y-axis: Profit
Each point = Order

Insight:

Some high-sales orders show low or negative profit.
Indicates:
High discounts
High costs
These orders should be investigated to improve profitability.

9. Chart Selection Justification

Explain why different charts were used for different business questions in this analysis.

- Different charts are used because:

Bar Chart: Best for comparing categories (clear comparison)
Line Chart: Shows trends over time
Pie Chart: Displays proportion/share
Scatter Plot: Identifies relationships (Sales vs Profit)
Heatmap: Shows patterns across two dimensions
Bubble Chart: Multi-variable analysis (Sales, Profit, Quantity)

 Each chart is chosen based on the type of analysis required (comparison, trend, distribution, relationship).
